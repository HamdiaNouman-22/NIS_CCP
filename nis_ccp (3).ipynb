```python
import string

class CombinedCipher:

    # Combined Vigenère + Affine Cipher

    def __init__(self):
        self.alphabet = string.ascii_uppercase
        self.M = 26

    def _gcd(self, a, b):
        while b:
            a, b = b, a % b
        return a

    def _prepare_text(self, text):
        return ''.join(c.upper() for c in text if c.isalpha())

    # Affine Implementation

    def affine_encrypt(self, plaintext, a, b):
        #Encrypting using Affine cipher: C = (aX + b) mod 26
        if self._gcd(a, self.M) != 1:
            raise ValueError(f"'a' must be coprime with {self.M}. Got a={a}")

        plaintext = self._prepare_text(plaintext)
        ciphertext = []
        for char in plaintext:
            p = self.alphabet.index(char)
            c = (a * p + b) % self.M
            ciphertext.append(self.alphabet[c])
        return ''.join(ciphertext)

    def affine_decrypt(self, ciphertext, a, b):
        #Decrypting using Affine cipher: P = (a^-1 * (Y - b)) mod 26
        if self._gcd(a, self.M) != 1:
            raise ValueError(f"'a' must be coprime with {self.M}. Got a={a}")

        ciphertext = self._prepare_text(ciphertext)
        a_inv = pow(a, -1, self.M)
        plaintext = []
        for char in ciphertext:
            c = self.alphabet.index(char)
            p = (a_inv * (c - b)) % self.M
            plaintext.append(self.alphabet[p])
        return ''.join(plaintext)

    # Vigenère Implementation

    def vigenere_encrypt(self, plaintext, key):
        # Encrypting using Vigenère cipher: C = (P + K[i]) mod 26
        plaintext = self._prepare_text(plaintext)
        key = self._prepare_text(key)
        ciphertext = []
        key_index = 0

        for char in plaintext:
            p = self.alphabet.index(char)
            k = self.alphabet.index(key[key_index % len(key)])
            c = (p + k) % self.M
            ciphertext.append(self.alphabet[c])
            key_index += 1
        return ''.join(ciphertext)

    def vigenere_decrypt(self, ciphertext, key):
        #Decrypting using Vigenère cipher: P = (C - K[i]) mod 26
        ciphertext = self._prepare_text(ciphertext)
        key = self._prepare_text(key)
        plaintext = []
        key_index = 0

        for char in ciphertext:
            c = self.alphabet.index(char)
            k = self.alphabet.index(key[key_index % len(key)])
            p = (c - k) % self.M
            plaintext.append(self.alphabet[p])
            key_index += 1
        return ''.join(plaintext)

    #  Combined Functions

    def combined_encrypt(self, plaintext, vigenere_key, affine_a, affine_b, order="VA"):

        #Combined encryption: Vigenère + Affine (VA) or Affine + Vigenère (AV).
        if order == "VA":
            temp = self.vigenere_encrypt(plaintext, vigenere_key)
            ciphertext = self.affine_encrypt(temp, affine_a, affine_b)
        else:  # AV
            temp = self.affine_encrypt(plaintext, affine_a, affine_b)
            ciphertext = self.vigenere_encrypt(temp, vigenere_key)
        return ciphertext

    def combined_decrypt(self, ciphertext, vigenere_key, affine_a, affine_b, order="VA"):

        #Combined Decryption: Reverses the encryption order.
        if order == "VA":
            # Decrypt: Affine inverse -> Vigenère inverse
            temp = self.affine_decrypt(ciphertext, affine_a, affine_b)
            plaintext = self.vigenere_decrypt(temp, vigenere_key)
        else:  # AV
            # Decrypt: Vigenère inverse -> Affine inverse
            temp = self.vigenere_decrypt(ciphertext, vigenere_key)
            plaintext = self.affine_decrypt(temp, affine_a, affine_b)
        return plaintext
```

```python
# CipherBreaker class for frequency attacks

import time
from collections import Counter

class CipherBreaker:

    def __init__(self, encryption_order="VA"):
        self.encryption_order = encryption_order
        self.alphabet = string.ascii_uppercase
        self.M = 26
        self.attack_log = []

        # English letter frequencies (in percentages)
        self.ENGLISH_FREQ = {
            'E': 12.70, 'T': 9.06, 'A': 8.17, 'O': 7.51, 'I': 6.97,
            'N': 6.75, 'S': 6.33, 'H': 6.09, 'R': 5.99, 'D': 4.25,
            'L': 4.03, 'C': 2.78, 'U': 2.76, 'M': 2.41, 'W': 2.36,
            'F': 2.23, 'G': 2.02, 'Y': 1.97, 'P': 1.93, 'B': 1.29,
            'V': 0.98, 'K': 0.77, 'J': 0.15, 'X': 0.15, 'Q': 0.10, 'Z': 0.07
        }

    def log(self, message):
        self.attack_log.append(message)
        print(message)

    def clear_log(self):
        self.attack_log = []

    def frequency_analysis(self, text):
        text = text.upper()
        total = sum(1 for c in text if c.isalpha())
        if total == 0: return {}
        freq = Counter(c for c in text if c.isalpha())
        return {letter: (count / total) * 100 for letter, count in freq.items()}

    def chi_squared(self, observed_freq):
        chi_sq = 0
        for letter in self.alphabet:
            expected = self.ENGLISH_FREQ.get(letter, 0)
            observed = observed_freq.get(letter, 0)
            if expected > 0:
                chi_sq += ((observed - expected) ** 2) / expected
        return chi_sq

    def index_of_coincidence(self, text):
        """Calculate Index of Coincidence for the text."""
        text = ''.join(c for c in text.upper() if c.isalpha())
        n = len(text)
        if n <= 1: return 0
        freq = Counter(text)
        ic = sum(f * (f - 1) for f in freq.values()) / (n * (n - 1))
        return ic

    def estimate_key_length(self, ciphertext, max_length=20):
        #Estimate Vigenère key length using Index of Coincidence.
        self.log("\nESTIMATING VIGENÈRE KEY LENGTH (IC Analysis)")
        ciphertext = ''.join(c for c in ciphertext.upper() if c.isalpha())
        ic_scores = {}
        for key_len in range(1, max_length + 1):
            groups = [''] * key_len
            for i, char in enumerate(ciphertext):
                groups[i % key_len] += char
            avg_ic = sum(self.index_of_coincidence(g) for g in groups) / key_len
            ic_scores[key_len] = avg_ic

        sorted_keys = sorted(ic_scores.items(), key=lambda x: abs(x[1] - 0.067))
        best_key_length = sorted_keys[0][0]
        self.log(f"Most likely key length: **{best_key_length}** (IC = {sorted_keys[0][1]:.4f})")
        return best_key_length, ic_scores
```

```python
def shift_cipher_encrypt(plaintext, shift):
    result = []
    for char in plaintext:
        if char.isalpha():
            result.append(chr(((ord(char.upper()) - 65 + shift) % 26) + 65))
        else:
            result.append(char)
    return ''.join(result)

def shift_cipher_decrypt(ciphertext, shift):
    return shift_cipher_encrypt(ciphertext, -shift)
```

```python
import math

# Helper: shift uppercase text by `shift` (positive shift = ROT forward). Non-alpha ignored.
def shift_text_upper(text, shift):
    out = []
    for ch in text:
        if 'A' <= ch <= 'Z':
            out.append(chr((ord(ch) - ord('A') + shift) % 26 + ord('A')))
    return ''.join(out)

def valid_affine_a_values():
    return [a for a in range(1, 26) if math.gcd(a, 26) == 1]

def guess_vigenere_key_for_blocks_by_chi(breaker, blocks):
    """
    For each block (letters in same key position), try all 26 shifts and pick the one
    whose shifted block has the lowest chi-squared score vs English.
    Returns the guessed key (string of letters).
    """
    guessed = []
    for block in blocks:
        best_shift = 0
        best_chi = float('inf')
        for k in range(26):
            candidate_plain_block = shift_text_upper(block, -k)
            obs_freq = breaker.frequency_analysis(candidate_plain_block)
            chi = breaker.chi_squared(obs_freq)
            if chi < best_chi:
                best_chi = chi
                best_shift = k
        guessed.append(chr(best_shift + ord('A')))
    return ''.join(guessed)

def va_frequency_attack(ciphertext, combined_cipher, breaker, max_key_len=16):
    """
    Try to break VA (Vigenere then Affine) ciphertext by:
      - Enumerating Affine (a,b), undoing affine first,
      - Estimating Vigenere key length on the result,
      - Recovering Vigenere key by per-block chi^2 shifts,
      - Scoring plaintext and returning best candidate.
    Returns dictionary with best candidate and metadata.
    """
    ciphertext_alpha = ''.join(c for c in ciphertext.upper() if c.isalpha())

    best = {
        'score': float('inf'),
        'plaintext': None,
        'vigenere_key': None,
        'affine_a': None,
        'affine_b': None,
        'chi': None
    }

    for a in valid_affine_a_values():
        for b in range(26):
            # Undo affine first (since encryption was Vigenere -> Affine)
            try:
                after_affine = combined_cipher.affine_decrypt(ciphertext_alpha, a, b)
            except ValueError:
                continue

            est_len, ic_scores = breaker.estimate_key_length(after_affine, max_length=min(max_key_len, len(after_affine) or 1))
            candidate_lengths = list(range(max(1, est_len-1), est_len+3))
            candidate_lengths = [L for L in candidate_lengths if L >= 1 and L <= min(max_key_len, len(after_affine) if len(after_affine)>0 else 1)]

            for L in candidate_lengths:
                blocks = [''] * L
                for i, ch in enumerate(after_affine):
                    blocks[i % L] += ch
                guessed_key = guess_vigenere_key_for_blocks_by_chi(breaker, blocks)
                v_decrypted = combined_cipher.vigenere_decrypt(after_affine, guessed_key)
                obs_freq_full = breaker.frequency_analysis(v_decrypted)
                chi_full = breaker.chi_squared(obs_freq_full)

                common_words = sum(v_decrypted.count(w) for w in ["THE", "AND", "TO", "OF", "IS", "IN", "YOU"])
                score = chi_full - (common_words * 6.0)

                if score < best['score']:
                    best.update({
                        'score': score,
                        'plaintext': v_decrypted,
                        'vigenere_key': guessed_key,
                        'affine_a': a,
                        'affine_b': b,
                        'chi': chi_full
                    })
    return best
def run_va_attack_improved():
    cipher = CombinedCipher()
    breaker = CipherBreaker(encryption_order="VA")

    va_test_cases = [
        {'name': 'Test Case 1', 'plaintext': 'HELLO', 'vigenere_key': 'SECRET', 'affine_a': 5, 'affine_b': 8},
        {'name': 'Test Case 2', 'plaintext': 'THE QUICK BROWN FOX JUMPS OVER THE LAZY DOG', 'vigenere_key': 'CRYPTOGRAPHYSECRET', 'affine_a': 5, 'affine_b': 8},
        {'name': 'Test Case 3', 'plaintext': 'NETWORK AND INFORMATION SECURITY', 'vigenere_key': 'PROJECTNIS486KEY', 'affine_a': 9, 'affine_b': 15}
    ]

    for test in va_test_cases:
        ciphertext = cipher.combined_encrypt(test['plaintext'], test['vigenere_key'], test['affine_a'], test['affine_b'], order="VA")
        print("\n=== Running attack on", test['name'], "===\nCiphertext:", ciphertext)
        result = va_frequency_attack(ciphertext, cipher, breaker, max_key_len=16)

        print("Best candidate plaintext:", result['plaintext'])
        print("Guessed Vigenere key:", result['vigenere_key'])
        print("Guessed Affine a,b:", (result['affine_a'], result['affine_b']))
        expected_plain_alpha = ''.join(c for c in test['plaintext'].upper() if c.isalpha())
        success = (result['plaintext'].replace(" ", "") == expected_plain_alpha)
        print(f"{test['name']} - {'✓ PASS' if success else '✗ FAIL'}\n")

if __name__ == "__main__":
    run_va_attack_improved()
```

```python
import math

# --- Helpers ---
def valid_affine_a_values():
    """Return all a in 1..25 that are coprime with 26 (valid affine a values)."""
    return [a for a in range(1, 26) if math.gcd(a, 26) == 1]

def shift_text_upper(text, shift):
    """Shift uppercase letters in text by 'shift' (positive = forward). Non-alpha removed."""
    out = []
    for ch in text:
        if 'A' <= ch <= 'Z':
            out.append(chr((ord(ch) - ord('A') + shift) % 26 + ord('A')))
    return ''.join(out)

def guess_vigenere_key_by_chi_on_ciphertext(breaker, ciphertext, key_len):
    """
    Build blocks from ciphertext (which is the Vigenere output in AV)
    and determine the key letter for each block by trying all 26 shifts
    and choosing the shift that minimizes chi-squared score for the
    candidate plaintext block.
    Returns guessed key as string.
    """
    # Build blocks
    blocks = [''] * key_len
    for i, ch in enumerate(ciphertext):
        if 'A' <= ch <= 'Z':
            blocks[i % key_len] += ch

    guessed = []
    for block in blocks:
        best_k = 0
        best_chi = float('inf')
        for k in range(26):
            candidate_plain_block = shift_text_upper(block, -k)  # plaintext candidate if key letter = k
            obs_freq = breaker.frequency_analysis(candidate_plain_block)
            chi = breaker.chi_squared(obs_freq)
            if chi < best_chi:
                best_chi = chi
                best_k = k
        guessed.append(chr(best_k + ord('A')))
    return ''.join(guessed)

def score_plaintext(breaker, plaintext):
    """Score plaintext: lower is better. Use chi^2 and reward common words."""
    obs = breaker.frequency_analysis(plaintext)
    chi = breaker.chi_squared(obs)
    common_words = sum(plaintext.count(w) for w in ["THE", "AND", "TO", "OF", "IS", "IN", "YOU"])
    # combine: lower chi better, more common words reduce score (better)
    return chi - (common_words * 6.0), chi, common_words

# --- AV frequency attack ---
def av_frequency_attack(ciphertext, combined_cipher, breaker, max_key_len=16, keep_top_n=5):
    """
    Attack AV ciphertext (Affine -> Vigenere).
    Steps:
      1. Estimate Vigenere key length(s) on ciphertext (Vigenere is last layer).
      2. For each candidate key length, guess Vigenere key by chi-sq per block.
      3. For each guessed key, vigenere_decrypt(ciphertext, key) -> after_vig (this should equal Affine(cipher_plain)).
      4. Enumerate all affine (a,b) and affine_decrypt(after_vig, a, b) to get candidate plaintexts.
      5. Score plaintexts and keep top N candidates.
    Returns list of top candidates sorted by score (ascending).
    """
    # Normalize ciphertext to uppercase letters only (your CombinedCipher expects that)
    ct_alpha = ''.join(c for c in ciphertext.upper() if c.isalpha())
    if not ct_alpha:
        return []

    # Step A: estimate best key length using IC, then try neighbors
    est_len, ic_scores = breaker.estimate_key_length(ct_alpha, max_length=min(max_key_len, len(ct_alpha)))
    candidate_lengths = list(range(max(1, est_len - 2), est_len + 3))
    # restrict to sensible bounds
    candidate_lengths = [L for L in candidate_lengths if 1 <= L <= min(max_key_len, max(1, len(ct_alpha)))]

    candidates = []  # will hold dicts with score, plaintext, keys

    # Step B: for each candidate key length guess the Vigenere key using chi^2 per block
    for L in candidate_lengths:
        guessed_key = guess_vigenere_key_by_chi_on_ciphertext(breaker, ct_alpha, L)

        # Step C: decrypt ciphertext with guessed Vigenere key (undo last layer)
        after_vig = combined_cipher.vigenere_decrypt(ct_alpha, guessed_key)

        # Step D: enumerate affine keys (a,b) and undo Affine
        for a in valid_affine_a_values():
            for b in range(26):
                try:
                    candidate_plain = combined_cipher.affine_decrypt(after_vig, a, b)
                except ValueError:
                    continue

                # Step E: score plaintext
                score, chi, common = score_plaintext(breaker, candidate_plain)

                candidates.append({
                    'score': score,
                    'chi': chi,
                    'common_words': common,
                    'plaintext': candidate_plain,
                    'vigenere_key': guessed_key,
                    'affine_a': a,
                    'affine_b': b,
                    'key_length_tried': L
                })

    # Sort by score (lower is better) and return top N
    candidates_sorted = sorted(candidates, key=lambda x: x['score'])
    return candidates_sorted[:keep_top_n]

# --- Runner similar to your tests ---
def run_av_attack_improved():
    cipher = CombinedCipher()
    breaker = CipherBreaker(encryption_order="AV")

    av_test_cases = [
        {'name': 'Test Case 1', 'plaintext': 'HELLO', 'vigenere_key': 'SECRET', 'affine_a': 5, 'affine_b': 8},
        {'name': 'Test Case 2', 'plaintext': 'THE QUICK BROWN FOX JUMPS OVER THE LAZY DOG', 'vigenere_key': 'CRYPTOGRAPHYSECRET', 'affine_a': 5, 'affine_b': 8},
        {'name': 'Test Case 3', 'plaintext': 'NETWORK AND INFORMATION SECURITY', 'vigenere_key': 'PROJECTNIS486KEY', 'affine_a': 9, 'affine_b': 15}
    ]

    for test in av_test_cases:
        ciphertext = cipher.combined_encrypt(test['plaintext'], test['vigenere_key'], test['affine_a'], test['affine_b'], order="AV")
        print("\n=== Running AV attack on:", test['name'], "===\nCiphertext:", ciphertext)

        top_candidates = av_frequency_attack(ciphertext, cipher, breaker, max_key_len=16, keep_top_n=5)

        if not top_candidates:
            print("No candidates generated.")
            continue

        print(f"Top {len(top_candidates)} candidates:")
        for i, cand in enumerate(top_candidates, 1):
            print(f"\nCandidate #{i}:")
            print("  Plaintext   :", cand['plaintext'])
            print("  Vigenere key:", cand['vigenere_key'], "(len:", cand['key_length_tried'], ")")
            print("  Affine a,b  :", (cand['affine_a'], cand['affine_b']))
            print("  Score (lower better):", cand['score'])
            print("  Chi-sq:", cand['chi'], "Common words:", cand['common_words'])

        # quick success check against expected plaintext (normalize both)
        exp_alpha = ''.join(c for c in test['plaintext'].upper() if c.isalpha())
        found = any(cand['plaintext'].replace(" ", "") == exp_alpha for cand in top_candidates)
        print(f"\n{test['name']} - {'✓ PASS' if found else '✗ FAIL'}")

# run when module executed directly
if __name__ == "__main__":
    run_av_attack_improved()
```

```python
import time

def run_tests():
    cipher = CombinedCipher()

    # Test configurations with varying plaintext lengths
    test_cases = [
        {
            'name': 'Test Case 1: VA Order (Vigenère -> Affine) - Short Length',
            'plaintext': 'HELLO',
            'vigenere_key': 'SECRET',
            'affine_a': 5,
            'affine_b': 8,
            'order': 'VA'
        },
        {
            'name': 'Test Case 2: VA Order (Vigenère -> Affine) - Medium Length',
            'plaintext': 'THE QUICK BROWN FOX JUMPS OVER THE LAZY DOG',
            'vigenere_key': 'CRYPTOGRAPHYSECRET',
            'affine_a': 5,
            'affine_b': 8,
            'order': 'VA'
        },
        {
            'name': 'Test Case 3: AV Order (Affine -> Vigenère) - Medium Length',
            'plaintext': 'NETWORK AND INFORMATION SECURITY IS A CRITICAL FIELD',
            'vigenere_key': 'PROJECTNIS486KEY',
            'affine_a': 9,
            'affine_b': 15,
            'order': 'AV'
        },
        {
            'name': 'Test Case 4: VA Order (Vigenère -> Affine) - Long Length',
            'plaintext': 'A' * 1000,  # 1000 character string
            'vigenere_key': 'LONGKEYLONGKEY',
            'affine_a': 7,
            'affine_b': 3,
            'order': 'VA'
        },
        {
            'name': 'Test Case 5: AV Order (Affine -> Vigenère) - Long Length',
            'plaintext': 'B' * 1000,  # 1000 character string
            'vigenere_key': 'LONGKEYLONGKEY',
            'affine_a': 11,
            'affine_b': 20,
            'order': 'AV'
        }
    ]

    # Shift Cipher test configuration
    shift_key = 3

    results_summary = []

    # Test for Combined Cipher
    for test in test_cases:
        total_time = 0
        iterations = 10

        for _ in range(iterations):
            start_time = time.time()

            # Encrypt the plaintext
            ciphertext = cipher.combined_encrypt(test['plaintext'], test['vigenere_key'], test['affine_a'], test['affine_b'], test['order'])

            # Decrypt the ciphertext
            decrypted_text = cipher.combined_decrypt(ciphertext, test['vigenere_key'], test['affine_a'], test['affine_b'], test['order'])

            # Success Check (Comparing decrypted text to the original plaintext)
            success = decrypted_text.replace(" ", "") == test['plaintext'].replace(" ", "")

            elapsed_time = time.time() - start_time
            total_time += elapsed_time

        avg_time = total_time / iterations
        results_summary.append({
            'test_name': test['name'],
            'success': '✓ PASS' if success else '✗ FAIL',
            'avg_time_taken': avg_time,
            'plaintext_length': len(test['plaintext'])
        })

        print(f"{test['name']} - {'✓ PASS' if success else '✗ FAIL'} | Average time taken: {avg_time:.6f} seconds | Plaintext Length: {len(test['plaintext'])}")

    # Test for Shift Cipher (Caesar Cipher)
    for test in test_cases:
        total_time = 0
        iterations = 10

        for _ in range(iterations):
            start_time = time.time()

            # Encrypt with Shift Cipher
            shift_ciphertext = shift_cipher_encrypt(test['plaintext'], shift_key)
            # Decrypt with Shift Cipher
            shift_decrypted_text = shift_cipher_decrypt(shift_ciphertext, shift_key)

            # Success Check (Comparing decrypted text to the original plaintext)
            shift_success = shift_decrypted_text.replace(" ", "") == test['plaintext'].replace(" ", "")

            shift_elapsed_time = time.time() - start_time
            total_time += shift_elapsed_time

        avg_shift_time = total_time / iterations
        results_summary.append({
            'test_name': f"Shift Cipher - {test['name']}",
            'success': '✓ PASS' if shift_success else '✗ FAIL',
            'avg_time_taken': avg_shift_time,
            'plaintext_length': len(test['plaintext'])
        })

        print(f"Shift Cipher - {test['name']} - {'✓ PASS' if shift_success else '✗ FAIL'} | Average time taken: {avg_shift_time:.6f} seconds | Plaintext Length: {len(test['plaintext'])}")

    return results_summary

test_results = run_tests()
```
